```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load the dataset
file_path = r"C:\Users\sebar\Downloads\weather_forecast_data.csv"  # Replace with your file path
weather_data = pd.read_csv(file_path)

# Encoding the target variable
le = LabelEncoder()
weather_data['Rain'] = le.fit_transform(weather_data['Rain'])

# Splitting dataset into features and target variable
X = weather_data.drop('Rain', axis=1)
y = weather_data['Rain']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)





# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Logistic Regression Model
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
log_reg_pred = log_reg.predict(X_test_scaled)

# Evaluating Logistic Regression
print("Logistic Regression Metrics:")
print(f"  Accuracy: {accuracy_score(y_test, log_reg_pred):.2f}")
print(f"  Precision: {precision_score(y_test, log_reg_pred):.2f}")
print(f"  Recall: {recall_score(y_test, log_reg_pred):.2f}")
print(f"  F1 Score: {f1_score(y_test, log_reg_pred):.2f}")
print(f"  Confusion Matrix:\n{confusion_matrix(y_test, log_reg_pred)}")
```

```python
# Importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load the dataset
file_path = r"C:\Users\sebar\Downloads\weather_forecast_data.csv"  # Replace with your file path
weather_data = pd.read_csv(file_path)

# Encoding the target variable
le = LabelEncoder()
weather_data['Rain'] = le.fit_transform(weather_data['Rain'])

# Splitting dataset into features and target variable
X = weather_data.drop('Rain', axis=1)
y = weather_data['Rain']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)

# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Support Vector Classifier (SVC) Model
svc_clf = SVC(kernel='linear', probability=True)
svc_clf.fit(X_train_scaled, y_train)
svc_clf_pred = svc_clf.predict(X_test_scaled)

# Evaluating SVC
print("Support Vector Classifier Metrics:")
print(f"  Accuracy: {accuracy_score(y_test, svc_clf_pred):.2f}")
print(f"  Precision: {precision_score(y_test, svc_clf_pred):.2f}")
print(f"  Recall: {recall_score(y_test, svc_clf_pred):.2f}")
print(f"  F1 Score: {f1_score(y_test, svc_clf_pred):.2f}")
print(f"  Confusion Matrix:\n{confusion_matrix(y_test, svc_clf_pred)}")
```

```python
# Importing libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load the dataset
file_path = r"C:\Users\sebar\Downloads\weather_forecast_data.csv"  # Replace with your file path
weather_data = pd.read_csv(file_path)

# Encoding the target variable
le = LabelEncoder()
weather_data['Rain'] = le.fit_transform(weather_data['Rain'])

# Splitting dataset into features and target variable
X = weather_data.drop('Rain', axis=1)
y = weather_data['Rain']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# K-Nearest Neighbors Model
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train_scaled, y_train)
knn_pred = knn_clf.predict(X_test_scaled)

# Evaluating K-Nearest Neighbors
print("K-Nearest Neighbors Metrics:")
print(f"  Accuracy: {accuracy_score(y_test, knn_pred):.2f}")
print(f"  Precision: {precision_score(y_test, knn_pred):.2f}")
print(f"  Recall: {recall_score(y_test, knn_pred):.2f}")
print(f"  F1 Score: {f1_score(y_test, knn_pred):.2f}")
print(f"  Confusion Matrix:\n{confusion_matrix(y_test, knn_pred)}")
```

```python
# Importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load the dataset
file_path = r"C:\Users\sebar\Downloads\weather_forecast_data.csv"  # Replace with your file path
weather_data = pd.read_csv(file_path)

# Encoding the target variable
le = LabelEncoder()
weather_data['Rain'] = le.fit_transform(weather_data['Rain'])

# Splitting dataset into features and target variable
X = weather_data.drop('Rain', axis=1)
y = weather_data['Rain']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scaling features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Gradient Boosting Model
gb_clf = GradientBoostingClassifier(random_state=42)
gb_clf.fit(X_train_scaled, y_train)
gb_pred = gb_clf.predict(X_test_scaled)

# Evaluating Gradient Boosting
print("Gradient Boosting Metrics:")
print(f"  Accuracy: {accuracy_score(y_test, gb_pred):.2f}")
print(f"  Precision: {precision_score(y_test, gb_pred):.2f}")
print(f"  Recall: {recall_score(y_test, gb_pred):.2f}")
print(f"  F1 Score: {f1_score(y_test, gb_pred):.2f}")
print(f"  Confusion Matrix:\n{confusion_matrix(y_test, gb_pred)}")
```

```python
# Importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Load the dataset
file_path = r"C:\Users\sebar\Downloads\weather_forecast_data.csv"  # Replace with your file path
weather_data = pd.read_csv(file_path)

# Encoding the target variable
le = LabelEncoder()
weather_data['Rain'] = le.fit_transform(weather_data['Rain'])

# Splitting dataset into features and target variable
X = weather_data.drop('Rain', axis=1)
y = weather_data['Rain']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scaling features (optional for Naive Bayes, as it handles raw features)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Naive Bayes Model
nb_clf = GaussianNB()
nb_clf.fit(X_train_scaled, y_train)
nb_pred = nb_clf.predict(X_test_scaled)

# Evaluating Naive Bayes
print("Naive Bayes Metrics:")
print(f"  Accuracy: {accuracy_score(y_test, nb_pred):.2f}")
print(f"  Precision: {precision_score(y_test, nb_pred):.2f}")
print(f"  Recall: {recall_score(y_test, nb_pred):.2f}")
print(f"  F1 Score: {f1_score(y_test, nb_pred):.2f}")
print(f"  Confusion Matrix:\n{confusion_matrix(y_test, nb_pred)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\sebar\Downloads\weather_forecast_data.csv"  # Replace with your file path
weather_data = pd.read_csv(file_path)
weather_data
```

```python
# Plotting histograms for Temperature, Humidity, Wind Speed, and Pressure
plt.figure(figsize=(15, 10))
features = ['Temperature', 'Humidity', 'Wind_Speed', 'Pressure']

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(weather_data[feature], kde=True, bins=30, color='skyblue')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Pair plot with 'Rain' as hue
sns.pairplot(weather_data, hue='Rain', palette='coolwarm')
plt.suptitle('Pair Plot of Weather Data', y=1.02)
plt.show()
```

```python
# Box plot of Temperature and Humidity by Rain
plt.figure(figsize=(12, 5))

# Box plot for Temperature
plt.subplot(1, 2, 1)
sns.boxplot(x='Rain', y='Temperature', data=weather_data, palette='Set2')
plt.title('Temperature Distribution by Rain')

# Box plot for Humidity
plt.subplot(1, 2, 2)
sns.boxplot(x='Rain', y='Humidity', data=weather_data, palette='Set2')
plt.title('Humidity Distribution by Rain')

plt.tight_layout()
plt.show()

```

```python
# Scatter plot of Temperature vs Humidity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='Humidity', hue='Rain', data=weather_data, palette='coolwarm')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.legend(title='Rain')
plt.show()
```

```python
# Line plot of Temperature and Humidity over time (index)
plt.figure(figsize=(12, 6))

# Temperature line plot
plt.plot(weather_data['Temperature'], label='Temperature', color='tomato')
plt.plot(weather_data['Humidity'], label='Humidity', color='skyblue')

plt.title('Temperature and Humidity Over Time')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Count plot of Rainy vs Non-Rainy days
plt.figure(figsize=(6, 4))
sns.countplot(x='Rain', data=weather_data, palette='Set2')
plt.title('Count of Rainy vs Non-Rainy Days')
plt.xlabel('Rain')
plt.ylabel('Count')
plt.show()
```
